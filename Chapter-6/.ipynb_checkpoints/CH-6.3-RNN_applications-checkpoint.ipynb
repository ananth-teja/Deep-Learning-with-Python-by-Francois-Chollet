#### Preparing the dataset

Keep in mind that we have 1 datapoint for every 10 minutes. 

- **lookback** - 720 ---(Observations will go back 5 days)
- **steps** - 6 ---(Observations will be sampled at one data point per hour)
- **delay** - 144 ---(Targets will be 24 hours in future)

Preprocessing of data will involve 2 steps - 
1. Data Normalization. 
2. Python generator - takes current array of float data and yields batches of data from the recent past, along with a target temperature in the future. Because the samples in the dataset are highly redundant(sample N and sample N+1 will have most of their timesteps in common), it would be wasteful to allocate every sample. Instead we will generate the sample on the fly using the original data. 



````python
mean = float_data[:200000].mean(axis = 0)
float_data -= mean
std = float_data[:200000].std(axis=0)
float_data /= std
````

***Data Generator***

It yields a tuple (samples, target), where sampples is one batch of input data and targets is the corresponding target temperatures, The generator takes the following arguments. 

- **data** - The original array of floating point data (normalized)
- **lookbackk** - How many time steps back the input data should go.
- **delay** - How many tiemsteps in the future the target shoulf be. 
- **min_index and max_index** - Indices in the data array that delimit which timesteps to draw from. Thi s is useful for keeping a segment of the data for validation and another for testing.
- **shuffle** - whether to shuffle the data or draw them in chronological order.
- **batch_size** - number of samples per batch
- **step** - period, in timesteps, at which we sample data. 

````python
def generator(data, lookback, delay, min_index, max_index, shuffle = False, batch_size = 128, step = 6):
    ### Defines the end of data as max_index 
    if max_index is None:
        max_index = len(data) - delay - 1
    ### The i is used as end point for data. So
    i = min_index + lookback
    while 1:
        if shuffle:
            rows = np.random.randint(min_index+lookback, max_index, size = batch_size)
        else:
            if i+batch_size >= max_index:
                i = min_index + lookback
            rows = np.arange(i, min(i+batch_size, max_index))
            
            i+=len(rows)
        
        samples = np.zeros((len(rows), lookback // step, data.shape[-1]))
        targets = np.zeros((len(rows),))
        
        for j,row in enumerate(rows):
            
            ### lookback is subtracted as we added it earlier, also step is a 
            ### parameter of range function which will sample the indices
            
            indices = range(rows[j]-lookback, rows[j], step)
            samples[j] = data[indices]
            ### The target value from future
            targets[j] = data[rows[j] + delay][1]
        yield samples, targets
````

***Now we have an abstracted the funstionality of the generator. We will define the generator for all three datasets using the above abstract function.***

````python
lookback = 1440
step = 6 
delay = 144
batch_size = 128
````

````python
train_gen = generator(float_data, 
                     lookback = lookback, 
                     delay = delay, 
                     min_index = 0,
                     max_index = 200000,
                     shuffle = True,
                     step = step,
                     batch_size = batch_size)
````

````python
val_gen = generator(float_data, 
                     lookback = lookback, 
                     delay = delay, 
                     min_index = 200001,
                     max_index = 300000,
                     step = step,
                     batch_size = batch_size)
````

````python
test_gen = generator(float_data, 
                     lookback = lookback, 
                     delay = delay, 
                     min_index = 300001,
                     max_index = None,
                     step = step,
                     batch_size = batch_size)
````

***Common Sense - Baseline SETUP***

Consider the problem of classification where the dataset is highly skewed (A:90 to B:10). Predicting just the class A would give us an accuracy of 90 percent which can be considered as simple baseline. Sometimes even this baseline is very difficult to beat even with complex ML algorithms. 

So, for this problem we can setup the baseline by predicting the next day temperature same as today's temperature which is not a very bad estimate. The temperature change is not drastic in a single day usually. So,  wee will measure the success of such a baseline model using Mean ABsolute Error.

````python
def evaluate_naive_method():
    batch_maes = []
    for step in range(val_steps):
        samples, targets = next(val_gen)
        preds = samples[:,-1,1]
        mae = np.mean(np.abs(preds - targets))
        batch_maes.append(mae)
    print(np.mean(batch_maes))
````

````python
evaluate_naive_method()
````

***Using Recurrent Dropouts to fight overfitting***

--------

In the research it has been proved that adding a Dropout layer before the RNN layer hinders learning rather than regularizing. So, ideally the same dropout mask should be applied at all the timesteps. The dropout masks should be applied to inner activations. This has been implemented and abstracted by the layers. It can be passed directly as arguments to the layer.

The two arguments related to dropout are - 
1. dropout : a float specifying dropout for input layers
2. recurrent_dropout : dropout of recurrent units

***The validation set is taking a long time to process so using 1/10th of data***

````python
val_gen = generator(float_data, 
                     lookback = lookback, 
                     delay = delay, 
                     min_index = 200001,
                     max_index = 210000,
                     step = step,
                     batch_size = batch_size)
````

***Bidirectional RNNs***

-----------

It is one of the varients of RNN, which sometimes offers a good improvement in accuracy over simple RNNs. It is more commonly used in NLP tasks. 

A Bidirectional RNN consists of two regular RNNs, such as GRU, and LSTM layers each of which processes the input in one direction. (chronologically and antichronologically) and then merges their representations. It sometimes helps understand patterns which a unidirectional RNN might miss. 

Now, we need to modify the generator a little bit for this. The reverse generator can be creted by just changing the yield statement to 
```python
yield samples[:,::-1,:], targets
```

-----

***Training and evaluating an LSTM using reversed sequences***

----

````python
from keras.datasets import imdb
from keras.preprocessing import sequence
from keras import layers
from keras.models import Sequential
````

````python
max_features = 10000
maxlen = 500
````

````python
(x_train, y_train), (x_test, y_test) = imdb.load_data()
````

````python
### Reversing sequences

x_train = [x[::-1] for x in x_train]
x_test = [x[::-1] for x in x_test]
````

````python
x_train = sequence.pad_sequences(x_train, maxlen = maxlen)
x_test = sequence.pad_sequences(x_test, maxlen = maxlen)
````

````python
model = Sequential()
model.add(layers.Embedding(max_features, 128))
model.add(layers.LSTM(32))
model.add(layers.Dense(1, activation = 'sigmoid'))
````

````python
model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['acc'])
````

````python
model.summary()
````

````python
history = model.fit(x_train, y_train, epochs = 10, batch_size = 128, validation_split = 0.2)
````

-------------

***Training and evaluating a Bidirectional LSTM***

-----------



````python
model = Sequential()
model.add(layers.Embedding(max_features, 128))
model.add(layers.Bidirectional(layers.LSTM(32)))
model.add(layers.Dense(1, activation = 'sigmoid'))
````

````python
model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['acc'])
````

````python
model.summary()
````

````python
history = model.fit(x_train, y_train, epochs = 10, batch_size = 128, validation_split = 0.2)
````

***The model performance is slightly better than simple LSTM.***
